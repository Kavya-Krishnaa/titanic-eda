```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv(r"C:\Users\smart\OneDrive\Desktop\titanic-datacleaning\cleaned_titanic.csv")
print(df.describe())

df.hist(figsize=(12,8),bins=20,edgecolor='black')
plt.suptitle("Histogram of numerical features")
plt.tight_layout()
plt.show()

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.boxplot(y=df['Age'])
plt.title("Boxplot:Age")

plt.subplot(1,2,2)
sns.boxplot(y=df['Fare'])
plt.title("Boxplot:Fare")

plt.tight_layout()
plt.show()

numeric_cols = ['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'Sex_male', 'Embarked_Q', 'Embarked_S']
plt.figure(figsize=(10,6))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()


sns.pairplot(df[['Survived','Age','Fare','Pclass','Sex_male']],hue='Survived')
plt.suptitle("Pairplot of selected features",y=1.02)
plt.show()














```
